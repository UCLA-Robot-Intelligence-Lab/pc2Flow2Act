```python
# robomimic dataset description https://robomimic.github.io/docs/datasets/overview.html
import numpy as np
import h5py
from tqdm import tqdm
import cv2
from matplotlib import pyplot as plt

dataset_path = "/data/yuanhong/mimicgen_data/core_datasets/coffee/demo_src_coffee_task_D1/demo_3d.hdf5"
demo_video_path = "/data/yuanhong/mimicgen_data/core_datasets/coffee/demo_src_coffee_task_D1/videos"

with h5py.File(dataset_path, 'r') as f:
    for i in range(50):
        image = f[f"data/demo_{i}/obs/agentview_image"]
        fps = 30
        frames = np.array(image)
        height, width = frames.shape[1:3]
        
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        video_writer = cv2.VideoWriter(f'{demo_video_path}/demo_{i}.mp4', fourcc, fps, (width, height))
        
        for frame in tqdm(frames):
            # Convert from RGB to BGR if needed:
            frame_bgr = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
            video_writer.write(frame_bgr)
        
        video_writer.release()
```
